```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
import seaborn as sns
from sklearn.model_selection import train_test_split
import sklearn
```

```python
df=pd.read_csv("car_data.csv")
df.head()
```

```python
data = pd.DataFrame(data={'x': df['Present_Price'], 'y': df['Selling_Price']})

# Create lmplot
lm = sns.lmplot('x', 'y', data, size=7, truncate=True, scatter_kws={"s": 10})

# Access the Figure
fig = lm.fig 

# Add a title to the Figure
fig.suptitle('Present Price vs Selling Price ', fontsize=12)
# Show the plot
plt.show()
```

```python
data = pd.DataFrame(data={'x': ((df['Present_Price']-df['Selling_Price'])/df['Present_Price']), 'y': df['Kms_Driven']})

# Create lmplot
lm = sns.lmplot('x', 'y', data, size=7, truncate=True, scatter_kws={"s": 60})

# Access the Figure
fig = lm.fig 

# Add a title to the Figure
fig.suptitle('Price Difference vs Kilometer Driven ', fontsize=12)
# Show the plot
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split( df[['Present_Price','Kms_Driven']],df['Selling_Price'] , test_size=0.33, random_state=42)
reg=linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
a = reg.predict(X_test)
```

```python
sklearn.metrics.mean_squared_error(y_test, a, sample_weight=None, multioutput='uniform_average', squared=True)
```

```python
sklearn.metrics.mean_absolute_error(y_test, a, sample_weight=None, multioutput='uniform_average')
```
